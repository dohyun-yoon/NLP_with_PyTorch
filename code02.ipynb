# 2-1. 텍스트 토큰화

```python
import spacy

nlp = spacy.load("en")
text = "Mary, don't slap the green witch"

[str(token) for token in nlp(text.lower())]
```

```python
from nltk.tokenize import TweetTokenizer

tweet = u"Show White and the Seven Degrees #MakeAMovieCold@midnight:-)"
tokenizer = TweetTokenizer()

tokenizer.tokenize(tweet.lower())
```

# 2-2. n-그램

```python
def n_grams(text, n):
    # takes tokens or text, returns a list of n-grams

    return [text[i:i+n] for i in range(len(text)-n+1)]

cleaned = ['mary', ',', 'do', "n't", 'slap', 'the', 'green', 'witch']

print(n_grams(cleaned, 3))
```

# 2-3. 표제어 추출

```python
import spacy

nlp = spacy.load('en')
doc = nlp(u"he was running late")
for token in doc:
    print('{} --> {}'.format(token, token.lemma_))
```

# 2-5. 품사 태깅

```python
import spacy

nlp = spacy.load("en")
doc = nlp(u"Mary slapped the green witch.")
for token in doc:
    print('{} --> {}'.format(token, token.pos_))
```

# 2-6. 청크 나누기(부분 구문 분석)

```python
import spacy

nlp = spacy.load("en")
doc = nlp(u"Mary slapped the green witch.")
for chunk in doc.noun_chunks: # 명사구(NP) 분석
    print('{} - {}'.format(chunk, chunk.label_))
```
